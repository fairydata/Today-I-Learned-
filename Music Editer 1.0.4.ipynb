```python
import os 
import glob
import eyed3 
from tkinter import * 
# 작업에 필요한 모듈을 불러옵니다.

def save(): 
    try: # 예외 처리
        import tkinter
        from tkinter import filedialog # 원하는 디렉토리를 선택합니다.
        for filename in glob.glob('*.mp3'):
            print(filename) # 
        
        selection = listNodes.curselection()
        if(len(selection)==0) :
            return
        
        name = tag1.cget("text") # 정보를 얻어옵니다. 
        
        title = tag2.get() 
        album = tag3.get() 
        artist = tag4.get() 
        
        print(name+"/"+label11.cget("text"))
        
        tag = eyed3.load(name+"/"+label11.cget("text"))
        tag.tag.album = album
        tag.tag.title = title 
        tag.tag.artist = artist 
        tag.rename(artist+' - '+title) # 파일명 수정 형태 지정
        tag.tag.save() 
        
        
        tag5.configure(text="성공적으로 변경되었습니다.") #성공시 메시지 출력
        
        
        root = tkinter.Tk()
        root.withdraw()
        dir_path = name
        listNodes.delete(0,END)
        file_list = os.listdir(dir_path) # 디렉토리 내의 전체 파일 불러오기
        file_list_py = [file for file in file_list if file.endswith(".mp3")] # 디렉토리 내에서 mp3 확장자만 모으기
        for filename in file_list_py:
            listNodes.insert(END, str(filename))

    except Exception as e:
        tag5.configure(text='에러가 발생했습니다. : '+str(e)) #에러 발생 시 알림 메시지 출력
        
def find():
    try:
        import tkinter
        from tkinter import filedialog

        root = tkinter.Tk()
        root.withdraw()
        dir_path = filedialog.askdirectory(parent=root,initialdir="/",title='음악 폴더 선택')
        tag1.configure(text=dir_path)
        tag5.configure(text="파일찾기")
        listNodes.delete(0,END)
        file_list = os.listdir(dir_path)
        file_list_py = [file for file in file_list if file.endswith(".mp3")]
        for filename in file_list_py:
            listNodes.insert(END, str(filename))

    except Exception as e:
        tag5.configure(text='파일찾기 실패 : '+str(e)) #실패 경우 메시지 출력
        
def click_item(event): 
    selection = listNodes.curselection()
    if(len(selection)==0) :
        return 
    
    tag2.delete(0, len(tag2.get())) 
    tag3.delete(0, len(tag3.get())) 
    tag4.delete(0, len(tag4.get())) 
    file_list = os.listdir(tag1.cget("text"))
    file_list_py = [file for file in file_list if file.endswith(".mp3")]
    label11.configure(text=file_list_py[selection[0]])

      
        
window = Tk()
window.geometry('800x280')
window.resizable(width=False, height=False) 


tag1 = Label(window, text = '')
tag5 = Label(window, text = '')
tag1.place(x=110, y=100) 
tag5.place(x=185, y=240)


label5 = Label(window, text = '실행 결과: ') 

label5.place(x=120, y=240)

tag2 = Entry(window) 
tag3 = Entry(window) 
tag4 = Entry(window)
label11=Label(window, text = '') 
label11.place(x=550, y=100)
tag2.place(x=550, y=130) 
tag3.place(x=550, y=160) 
tag4.place(x=550, y=190) 
label2 = Label(window, text = '곡 이름 입력: ') 
label3 = Label(window, text = '앨범 이름 입력: ') 
label4 = Label(window, text = '아티스트 입력: ') 
label10= Label(window, text = '선택 파일 : ') 
label2.place(x=450, y=130) 
label3.place(x=450, y=160) 
label4.place(x=450, y=190)
label10.place(x=450, y=100)

label7 = Label(window, text = '파이썬을 이용한 음악 파일명 변경 프로그램 ver 1.0.4', fg = 'black', font = 'Gothic 20 bold')
label6 = Label(window, text = '광운대 21-1학기 컴퓨터언어 기말과제 2019510035 백지헌', fg = 'black', font = 'Gothic 10 italic')
label8 = Label(window, text = '변경하고자 하는 파일이 자신의 지정 디렉토리에 있어야 합니다.', fg = 'green')
label7.place(x=60, y=25)
label6.place(x=200, y=65)
label8.place(x=40, y=130)

button1 = Button(window, text = "폴더찾기", command=find) 
button3 = Button(window, text = "수정 적용", command=save) 
button1.place(x=40, y=97) 
button3.place(x=630, y=220)

frame =Frame(window)
frame.pack()

listNodes = Listbox(window, width=40, height=5)
listNodes.place(x=40, y=150)

listNodes.bind('<ButtonRelease-1>', click_item)


window.mainloop()
```
